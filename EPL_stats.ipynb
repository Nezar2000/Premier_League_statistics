```python
# Import the libraries


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load Dataset

epl_df = pd.read_csv('EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
# Create minutes and goals per match columns


epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float).round(2)
epl_df.head()
```

```python
# Total Goals 


Total_goals = epl_df['Goals'].sum()
print(Total_goals)
```

```python
# Penalty Goals


Penalty_goals = epl_df['Penalty_Goals'].sum()
print(Penalty_goals)
```

```python
# Penalty attempts


Total_PenaltyAttemptes = epl_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttemptes)
```

```python
# Pie chart for penalties scored vs missed


plt.figure(figsize =(7,3))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - Total_Penaltygoals
data = [pl_not_scored, Total_Penaltygoals]
labels = ['Penalities missed','Penalties scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Most players from wich countries


nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("bright"))
```

```python
# Clubs with the maximum players in the squad


epl_df['Club'].value_counts().nlargest(7).plot(kind='bar', color=sns.color_palette("viridis"))
```

```python
# Clubs with the least players in the squad


epl_df['Club'].value_counts().nsmallest(7).plot(kind='bar', color=sns.color_palette("viridis"))
```

```python
# Players based on age group


under_20 = epl_df[epl_df['Age']<= 20]
age20_25 = epl_df[(epl_df['Age']>20) & (epl_df['Age']<=25)]
age25_30 = epl_df[(epl_df['Age']>25) & (epl_df['Age']<=30)]
above_30 = epl_df[(epl_df['Age']>30)]
```

```python
x = np.array([under_20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),above_30['Name'].count()])
mylabels = ["U 20","between 20 & 24","25 & 29","above 30"]
plt.title('Total players with age groups',fontsize = 15)
plt.pie(x, labels = mylabels, autopct = "%.1f%%")
plt.figure(figsize =(7,3))
plt.show()
```

```python
# Total players under 20 in each club


players_under_20 = epl_df[epl_df['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',color = sns.color_palette("cubehelix"))
```

```python
# Under 20 in ManUnited


players_under_20[players_under_20['Club']=='Manchester United']
```

```python
# Under 20 in Chelsea


players_under_20[players_under_20['Club']=='Chelsea']
```

```python
#Average age of players in each club


plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data = epl_df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Let's make sure


num_players = epl_df.groupby('Club').size()
data = epl_df.groupby('Club')['Age'].sum() / num_players
data.sort_values(ascending=False)
```

```python
# Total assists from each clubs


Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel("Club",fontsize=20)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(12,6)
plt.title('Plot of Clubs vs Total Assists',fontsize=30)
plt.show()
```

```python
# Top 10 assists


Top_10_Assists = epl_df[['Name','Club','Assists','Matches']].nlargest(10,columns='Assists')
Top_10_Assists
```

```python
# Total goals from each club


Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Club",fontsize=20)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(12,6)
plt.title('Plot of Clubs vs Total Goals',fontsize=30)
plt.show()
```

```python
# Most goals by players


Top_10_Goals = epl_df[['Name','Club','Goals','Matches']].nlargest(10,columns='Goals')
Top_10_Goals
```

```python
# Goals per match


top_10_goals_per_match = epl_df[['Name','GoalsPerMatch','Matches','Goals']].nlargest(n=10,columns='GoalsPerMatch')
top_10_goals_per_match
```

```python
# Pie chart goals with assists and without assists


plt.figure(figsize=(8,3))
assists=epl_df['Assists'].sum()
data = [Total_goals - assists, assists]
labels = ['Goals without assits','Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data,labels = labels, colors = color,autopct = '%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards


epl_yellow=epl_df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(12,6))
plt.title("Players with the most yellow cards")
c=sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'],label='Players',color='Yellow')
plt.ylabel('Number of yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
plt.show()
```
